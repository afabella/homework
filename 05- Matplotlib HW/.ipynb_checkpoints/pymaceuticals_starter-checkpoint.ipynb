```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)


# Combine the data into a single dataset

mouse_trial_complete = pd.merge(mouse_data, clinical_trial_data, on= "Mouse ID", how="left")
mouse_trial_complete = mouse_trial_complete[(mouse_trial_complete["Drug"] == 'Capomulin')|
                                           (mouse_trial_complete["Drug"] == 'Infubinol')|
                                           (mouse_trial_complete["Drug"] == 'Ketapril')|
                                           (mouse_trial_complete["Drug"] =='Placebo')]

# Display the data table for preview

mouse_trial_complete.head(100)

from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)


# Combine the data into a single dataset

mouse_trial_complete = pd.merge(mouse_data, clinical_trial_data, on= "Mouse ID", how="left")
mouse_trial_complete = mouse_trial_complete[(mouse_trial_complete["Drug"] == 'Capomulin')|
                                           (mouse_trial_complete["Drug"] == 'Infubinol')|
                                           (mouse_trial_complete["Drug"] == 'Ketapril')|
                                           (mouse_trial_complete["Drug"] =='Placebo')]

# Display the data table for preview

mouse_trial_complete.head(100)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

drug_timepoint_df = mouse_trial_complete.groupby(["Drug", "Timepoint"])
mean_tumor_volume_data = drug_timepoint_df["Tumor Volume (mm3)"].mean()
mean_volume_df = pd.DataFrame(
    {"Tumor Volume Mean": mean_tumor_volume_data
     },
      
)
mean_volume_df = mean_volume_df.reset_index()
mean_volume_df.head(20)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
standard_errors = drug_timepoint_df["Tumor Volume (mm3)"].sem()
standard_error_df = pd.DataFrame(
    {"Standard Errors": standard_errors     
     }
)
# standard_error_df = standard_error_df
standard_error_df.head(50)
```

```python
volume_plot = mean_volume_df.plot.scatter(x='Timepoint', y='Tumor Volume Mean', c = 'Red')
plt.title ('Tumor Volume Over Time', fontdict = {'fontsize' : 15})

#save the figure
plt.savefig('tumor_volume_over_time.png')
#Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met_data = drug_timepoint_df["Metastatic Sites"].mean()
# Convert to DataFrame
mean_meta_df = pd.DataFrame(
    {"Mean Met. Site": mean_met_data     
     }
)
mean_meta_df = mean_meta_df.reset_index()
# Preview DataFrame
mean_meta_df
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
standard_error_site = drug_timepoint_df["Metastatic Sites"].sem()
standard_error_site
# Convert to DataFrame
standard_site_df = pd.DataFrame({
    "Site Standard Error": standard_error_site
    }
)
# Preview DataFrame
standard_site_df
```

```python
# Generate the Plot (with Error Bars)
metastatic_plot = mean_meta_df.plot.scatter(x='Timepoint',
                      y='Mean Met. Site',
                      c='Red')
plt.title('Metastatic Spread During Treatment')

# Save the Figure
plt.savefig('metastatic_spread_during_treatment.png')
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_grouped_data = drug_timepoint_df["Mouse ID"].count()
# Convert to DataFrame
mouse_grouped_df = pd.DataFrame({
    "Mice Grouped": mouse_grouped_data
})
    
#Preview DataFrame
mouse_grouped_df = mouse_grouped_df.reset_index()
mouse_grouped_df.head(50)
```

```python
total_mice = mouse_grouped_df["Mice Grouped"].sum()
(mouse_grouped_data/total_mice)*100
```

```python
# Generate the Plot (Accounting for percentages)
mouse_survival_plot = mouse_grouped_df.plot.scatter(x='Timepoint', y='Mice Grouped',c='DarkBlue')
plt.title('Survival During Treatment')


# Save the Figure
plt.savefig('mouse_survival.png')
# Show the Figure
plt.show()
```
